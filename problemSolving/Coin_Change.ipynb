```python
def coin_change(coins: list[int], amount: int):
    """
    Given an integer array coins representing coins of different denominations and an integer amount representing a total amount of money, return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return -1.
    :param coins:  the denominations of the coins
    :param coins type: List[int]
    :returns:  the fewest number of coins needed to make up the amount, or -1 if it's not possible
    :rtype: int
    """
    dp = [amount + 1] * (amount + 1) # Initialize the dp list  with a value greater than the maximum possible number of coins (amount + 1)
    dp[0] = 0                        #0 coins are needed to make up the amount of 0
    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], 1 + dp[i - coin])
    return dp[amount] if dp[amount] <= amount  else -1

if __name__ == "__main__":
    coins = [1, 2, 5]
    amount = 11
    print(coin_change(coins, amount)) # Output: 3 (11 can be made with 5 + 5 + 1)
    coins = [2]
    amount = 3
    print(coin_change(coins, amount)) # Output: -1 (3 cannot be made with only 2)
    coins = [1]
    amount = 0
    print(coin_change(coins, amount))  # 
    
```

```python
def Basic_Calculator_II(s: str):
    """
    Evaluates a basic arithmetic expression containing integers and +, -, *, / operators.
    Args:
        s (str): Arithmetic expression.
    Returns:
        int: Result of the expression.
    """
    stack = []  # stack for numbers to handle addition/subtraction
    num = 0     # current number being built
    op = '+'    # previous operator

    for i in range(len(s)):
        if s[i].isdigit():
            num = num * 10 + int(s[i])  # build multi-digit number
        if s[i] in '+-*/' or i == len(s) - 1:
            if op == '+':
                stack.append(num)
            elif op == '-':
                stack.append(-num)
            elif op == '*':
                stack.append(stack.pop() * num)
            elif op == '/':
                stack.append(int(stack.pop() / num))  # truncate towards zero
            num = 0
            op = s[i]

    return sum(stack)

# Example usage

expressions = [
    "3+2*2",
    " 3/2 ",
    " 3+5 / 2 "
]

for exp in expressions:
    result = Basic_Calculator_II(exp)
    print(f"Input: s = '{exp}'")
    print(f"Output: {result}")
    print()
```

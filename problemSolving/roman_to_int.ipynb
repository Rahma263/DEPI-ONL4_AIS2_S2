```python
def roman_to_integer(s: str) -> int:
    """
    Convert a Roman numeral string to an integer.

    Args:
        s (str): Roman numeral string (e.g., "XIV", "MCMXCIV").

    Returns:
        int: The integer value of the Roman numeral.
    """
    # Mapping of Roman numerals to their integer values
    roman_to_int = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 
                    'C': 100, 'D': 500, 'M': 1000}
    
    total = 0       # Running total of the integer value
    prev_value = 0  # Value of the previous numeral (used for subtraction logic)
    
    # Iterate over the Roman numeral from right to left
    for char in reversed(s):
        value = roman_to_int[char]
        
        # If current numeral is smaller than previous, subtract it else, add it
        if value < prev_value:
            total -= value
        else:
            total += value
        
        # Update previous value for next iteration
        prev_value = value
    
    return total
print(roman_to_integer("III"))
print(roman_to_integer("LVIII"))
print(roman_to_integer("MCMXCIV"))
```

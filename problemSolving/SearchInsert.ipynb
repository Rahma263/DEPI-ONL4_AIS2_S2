```python
def search_insert_position(lst: list, target: int):
    """
    Find the index of the target in a sorted list using binary search.  
    If the target is not found, return -1.

    Args:
        lst (list): A list of integers sorted in ascending order.
        target (int): The integer value to search for.

    Returns:
        int: The index of the target if found; otherwise, -1.
    """
    start = 0
    end = len(lst) - 1

    # Standard binary search loop
    while start <= end:
        mid = (start + end) // 2  # Find the middle index in int

        if lst[mid] == target:
            return mid  # Target found, return index
        elif lst[mid] < target:
            start = mid + 1  # Search in the right half
        else:
            end = mid - 1    # Search in the left half

    # Target not found
    return -1


# Example usage
print(search_insert_position([1, 3, 5, 6], 5))  
print(search_insert_position([1, 3, 5, 6], 2))  
print(search_insert_position([1, 3, 5, 6], 7))  
```

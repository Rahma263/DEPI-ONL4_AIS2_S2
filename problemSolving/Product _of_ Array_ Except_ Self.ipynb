```python
def product_of_array_except_self(nums: list) :
    """Returns an array where each element is the product of all the numbers in the input array except the one at that index.
    nums: List[int] - The input array of integers.
    returns: List[int] - An array where each element is the product of all the numbers in the input array except the one at that index.
    """
    
    n = len(nums)
    output = []
    
    for i in range(n):
        product = 1
        for j in range(n):
            if i != j:
                product *= nums[j]
        output.append(product)
    
    return output
nums = [1,2,3,4]
print(product_of_array_except_self(nums))

nums = [-1,1,0,-3,3]
print(product_of_array_except_self(nums))

    
```

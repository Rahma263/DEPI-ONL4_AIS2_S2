```python
from collections import deque

class TreeNode:
    """
    Represents a single node in a binary tree.
    Each node holds a value and references to its left and right children.
    """
    def __init__(self, value=0, left=None, right=None):
        self.value = value    # node's value
        self.left = left      # left child
        self.right = right    # right child


def maxDepth(nums):
    """
    Determines the maximum depth of a binary tree represented as a level-order list.

    Args:
        nums (list): Level-order list representation of a binary tree.
                     Use None for missing nodes. Example: [5, 1, 4, None, None, 3, 6]

    Returns:
        bool: True if the tree is a valid BST, False otherwise.
    """

    # empty list is  valid
    if not nums:
        return True

    # create root from first element
    root = TreeNode(nums[0])

    # queue holds (node, index) pairs for BFS tree building
    queue = deque([(root, 1)])
    i = 1

    # build tree level by level
    while queue and i < len(nums):
        node, _ = queue.popleft()

        # assign left then right child
        for side in ['left', 'right']:
            if i < len(nums) and nums[i] is not None:
                new_node = TreeNode(nums[i])
                setattr(node, side, new_node)   # node.left or node.right = new_node
                queue.append((new_node, i))     # add child to queue for next level
            i += 1
    def depth(node):
        """
        Recursively calculates depth at each node.

        Args:
            node (TreeNode): Current node.

        Returns:
            int: Depth of subtree rooted at this node.
        """
        # empty node contributes 0 depth
        if node is None:
            return 0

        # current node = 1 + whichever side is deeper
        return 1 + max(depth(node.left), depth(node.right))

    return depth(root)
        
    


# --- Tests ---
print(maxDepth([3, 9, 20, None, None, 15, 7]))  
print(maxDepth([1, None, 2]))                                                
print(maxDepth([1]))                             
```

```python
def addTwoNumber(l1: list, l2: list) -> list:
    """
    Adds two numbers represented as reversed lists of digits and returns the sum 
    as a reversed list of digits.

    Args:
        l1 (list): First number represented as a list of digits in reverse order.
        l2 (list): Second number represented as a list of digits in reverse order.

    Returns:
        list: Sum of the two numbers, represented as a list of digits in reverse order.
    """
    num1 = 0
    num2 = 0
    
    # Convert first list of digits to a number
    for i in range(len(l1)):
        num1 += l1[i] * 10**i
    
    # Convert second list of digits to a number
    for j in range(len(l2)):
        num2 += l2[j] * 10**j
    
    # Add the two numbers
    result = num1 + num2
    
    # Convert the sum to a reversed list of digits
    reversed_digit = [int(digit) for digit in str(result)[::-1]]
    
    return reversed_digit

# Examples
print(addTwoNumber([2,4,3], [5,6,4]))       # Output: [7,0,8]
print(addTwoNumber([9,9,9,9,9,9,9], [9,9,9,9]))  # Output: [8,9,9,9,0,0,0,1]
print(addTwoNumber([0], [0])) 
```

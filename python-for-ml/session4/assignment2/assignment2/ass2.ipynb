```python
import os
import random

def thanos_project(folder_path, num_files):
    """
    thanos project:
       make a function to do the following:
       1-open the whole folder
       2-create number of files inside it
       3-check the len of files in that folder
       4-delete random half of this folder
       5-check the result
       :param_1:folder_path:path to the folder
       :param_2:num_files:number of the filesto create
       :return:None

    """
    os.makedirs(folder_path, exist_ok=True)

    for i in range(num_files):
        with open(os.path.join(folder_path, f"file_{i}.txt"), 'w') as f:
            f.write(f"this is file number {i}")

    all_files = os.listdir(folder_path)
    print("Before deletion:", len(all_files))

    files_to_delete = random.sample(all_files, len(all_files)//2)

    for file_name in files_to_delete:
        os.remove(os.path.join(folder_path, file_name))

    remaining_files = os.listdir(folder_path)
    print("After deletion:", len(remaining_files))

thanos_project("thanos_folder", 10)
```

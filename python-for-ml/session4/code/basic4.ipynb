```python
def factorial_1(num :int):
    if num == 0:
        return 1
    if num == 1:
        return 1
    
    
    return num* factorial_1(num-1)
```

```python
factorial_1(5)
```

```python
def is_prime (num:int):
    for i in range(2, num):
        if num %i ==0:
            return False
    return True
```

```python
is_prime(3)
```

```python
def reverse_text(name:str):
    """
        this function take a name from user interface
        :param:the name of the user
        type param_1:str
        return:the name of the user in reversed
        rtype:str
    """
    split_txt=name.split(" ")
    new_list=split_txt[::-1]
    result=" ".join(new_list)
    return result
```

```python
reverse_text("welcome depi")
```

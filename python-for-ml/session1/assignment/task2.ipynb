```python
# Decode an encoded message
msg = "###!!@mocleW PGTQ!!!6789"

# Remove unwanted parts
clean_msg = msg.replace("###!!@", "").replace("!!!6789", "")

# Split the message into parts
substring = clean_msg.split(" ")

# Reverse the first part
substring[0] = substring[0][::-1]

# Define vowel shift mapping
shift_vowels = {"E": "A", "I": "E", "O": "I", "U": "O", "A": "U"}

# Split the second part into characters
to_findvowels = list(substring[1])  

#check if the second word has a vowels and to shift them
for i in range(len(to_findvowels)):
    if to_findvowels[i] in shift_vowels:
         for i in range(len(to_findvowels)):
            if to_findvowels[i] in shift_vowels:
                to_findvowels[i] = shift_vowels[to_findvowels[i]]
                 
    else:
        continue
# Combine the message
decoded_msg = " ".join(substring)
print(decoded_msg)    #the output will be welcome PLAY
```

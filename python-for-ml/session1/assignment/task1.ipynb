```python
# check if the email invaild or vaild
x= "Amit_ml@gmail.edu"
if x.count("@") ==1 and (x.endswith(".edu") or x.endswith(".com")):   #check for @ and .edu
    print("vaild email")
else:
    print("invaild email")
position=x.find("@")      #find the position of the @
username=x[0:position]     #get the username
dotposition=x.find(".")     # find the postion of .
domain=x[position+1:dotposition]   #get the domain name
if domain==".com":            #check for domain have .com 
    print("commercial domain")
else:                          #check for domain have .edu
    print("Educational domain") 
print(f"username:{username}, domain:{domain}")
```

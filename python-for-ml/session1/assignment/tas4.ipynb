```python
# Decode an encoded message
msg="##$$$@!yalptcejorp EPUVT****9887"
# Remove unwanted parts
cleaned_msg=msg.replace("##$$$@!","").replace("****9887","")
#split the cleaned message into words
words=cleaned_msg.split()
#reverse the first word
words[0]=words[0][::-1]
# Define vowel shift mapping
shift_vowels = {"E": "A", "U": "O"}
# Split the second part into characters
to_findvowels = list(words[1]) 
#  check if the word  have Shift vowels or not
for i in range(len(to_findvowels)):
    if to_findvowels[i] in shift_vowels:
        to_findvowels[i] = shift_vowels[to_findvowels[i]]
# Join back the second part
words[1] = "".join(to_findvowels)
# Combine the message
decoded_msg = " ".join(words)
print(decoded_msg)  #output:projactplay APOVT
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#This is the sample 2 dimensional data we are going to reduce
 
x = np.matrix([2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1])
y = np.matrix([2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9])
original_data = np.r_[x,y]
```

```python
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
mean_x = np,mean(x)
mean_y = np.mean(y)
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = np.r_[x,y]
cov = np.matmul(data,data.T)/(10-1)
print(cov)
```

```python
evals , evecs = np.linalg.eig(cov)
print(evals)
print(evecs)
```

```python
evecs = np.flip(evecs,axis=1)
evals = np.flip(evals)
print(evecs)
print(evals)
```

```python
print(np.dot(evecs[:,0].T,original_data))
```

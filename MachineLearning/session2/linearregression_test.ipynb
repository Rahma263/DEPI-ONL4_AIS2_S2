```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #function
from sklearn.linear_model import LinearRegression  #class
from sklearn.metrics import mean_squared_error , r2_score,root_mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data, x_vars=["YearsExperience"], y_vars=["Salary"], height=7, kind='scatter')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Salary Prediction")
plt.show()
```

```python
x = data.iloc[:,:-1]  #split the x it is all columns except the last columns as it is a target
y = data.iloc[:,1]   # make the y is the last columns as it is target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , train_size=0.8, random_state = 10  ) #automatic he knows that the test_size = 0.2 vice versa
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train) #fit() is class abstract so it make the bkue table as an output
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train , linear_model.predict(x_train),color='blue')
plt.title("salary vs Experience if years")
plt.xlabel("years of Experience")
plt.ylabel("salary")
plt.show()  
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test , linear_model.predict(x_test),color='blue')
plt.title("salary vs Experience if years")
plt.xlabel("years of Experience")
plt.ylabel("salary")
plt.show()  
```

```python
y_test
```

```python
y_pred = linear_model.predict(x_test)
y_pred
```

```python
#calculate mean square error
mse = mean_squared_error(y_test,y_pred)
print(mse)
rms = root_mean_squared_error(y_test,y_pred)
print(rms)   # the gap that the model have made an error
r2score = r2_score(y_test,y_pred)
print(r2score)   # Not accuracy it is 1 - covarence of my model(wx+b=y)/covarence of the original line
```
